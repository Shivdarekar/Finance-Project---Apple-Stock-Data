## Data Visualization
 - Using Matplotlib & Seaborn visualization libraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# figure fuction manage the width and height of the histogram.
plt.figure(figsize =(7,5))

# Hist fuction is to create the histogram . When you pass the particular columns.
plt.hist(x = "Open" , data = data_clean , width = 10 , color = "c")

# Title show the title of histogram and X label & Y label show the label on X&Y axis.
plt.title("Stock Histogram Chart")
plt.xlabel("Open")
plt.ylabel("Close")
plt.show()
```

```python
groupA = data_clean["Open"].nunique()
groupB = data_clean["High"].nunique()
groupC = data_clean["Low"].nunique()
groupD = data_clean["Close"].nunique()


my_lables = ["Open","High","Low","Close"]
my_values = [groupA, groupB, groupC, groupD]

plt.pie(my_values, labels = my_lables, autopct= "%1.0f%%", explode = (0,0.05,0,0.1))
plt.title("Percent wise Stock Data")

# legend --> It shows elements names in the graphs.
plt.legend(title = "Climate Name",loc = "center")
plt.show()
```

```python
plt.figure(figsize =(12,5))
plt.plot(data_clean.Volume)
plt.xlabel("Open")
plt.ylabel("Close")
plt.title("Line chart")
plt.show()
```

```python
plt.figure(figsize =(12,5))
plt.scatter(data_clean.Open,data_clean.Open, color = "k",label = "DATA")
plt.xlabel("Open")
plt.ylabel("Close")
plt.legend(loc = "right")
plt.title("Scatter Plot 1")
plt.show()
```

```python
# Kdeplot show the single line graph.
sns.kdeplot(data_clean)
plt.show()
```
